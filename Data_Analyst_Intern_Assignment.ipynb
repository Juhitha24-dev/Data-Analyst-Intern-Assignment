DATA LOADING AND INSPECTION

```python
import pandas as pd

# Read the CSV files into Pandas DataFrames
users_df = pd.read_csv("UserDetails.csv")
sessions_df = pd.read_csv("CookingSessions.csv")
orders_df = pd.read_csv("OrderDetails.csv")

# Display the first 5 rows of each DataFrame
print("UserDetails DataFrame Head:")
print(users_df.head().to_markdown(index=False, numalign="left", stralign="left"))

print("\nCookingSessions DataFrame Head:")
print(sessions_df.head().to_markdown(index=False, numalign="left", stralign="left"))

print("\nOrderDetails DataFrame Head:")
print(orders_df.head().to_markdown(index=False, numalign="left", stralign="left"))

# Print the column names and their data types for each DataFrame
print("\nUserDetails DataFrame Info:")
print(users_df.info())

print("\nCookingSessions DataFrame Info:")
print(sessions_df.info())

print("\nOrderDetails DataFrame Info:")
print(orders_df.info())
```

```python
# Get all unique dishes from `sessions_df`
unique_dishes_sessions = sessions_df["Dish Name"].unique()

# Get all unique dishes from `orders_df`
unique_dishes_orders = orders_df["Dish Name"].unique()

# Check the number of unique dishes in `sessions_df`
if len(unique_dishes_sessions) > 50:
    # Calculate value counts for dishes in `sessions_df` and print the top 50 in a list
    print("Top 50 most occurring dishes in CookingSessions:")
    print(sessions_df["Dish Name"].value_counts().head(50).index.tolist())
else:`
    print("All unique dishes in CookingSessions:")
    print(unique_dishes_sessions)

# Check the number of unique dishes in `orders_df`
if len(unique_dishes_orders) > 50:
    print("Top 50 most occurring dishes in OrderDetails:")
    print(orders_df["Dish Name"].value_counts().head(50).index.tolist())
else:
    print("All unique dishes in OrderDetails:")
    print(unique_dishes_orders)
```

```python
# Drop rows with null values in all DataFrames
users_df.dropna(inplace=True)
sessions_df.dropna(inplace=True)
orders_df.dropna(inplace=True)

# Convert `Registration Date` column in `users_df` to datetime type
users_df["Registration Date"] = pd.to_datetime(users_df["Registration Date"])

# Convert `Session Start` and `Session End` columns in `sessions_df` to datetime type
sessions_df["Session Start"] = pd.to_datetime(sessions_df["Session Start"])
sessions_df["Session End"] = pd.to_datetime(sessions_df["Session End"])

# Convert `Order Date` column in `orders_df` to datetime type
orders_df["Order Date"] = pd.to_datetime(orders_df["Order Date"])

# Display the first 5 rows of each DataFrame
print("UserDetails DataFrame Head:")
print(users_df.head().to_markdown(index=False, numalign="left", stralign="left"))

print("\nCookingSessions DataFrame Head:")
print(sessions_df.head().to_markdown(index=False, numalign="left", stralign="left"))

print("\nOrderDetails DataFrame Head:")
print(orders_df.head().to_markdown(index=False, numalign="left", stralign="left"))

print("\nUserDetails DataFrame Info:")
print(users_df.info())

print("\nCookingSessions DataFrame Info:")
print(sessions_df.info())

print("\nOrderDetails DataFrame Info:")
print(orders_df.info())
```

```python
# Merge `sessions_df` with `orders_df` on `Session ID` and `User ID`
merged_df = pd.merge(
    sessions_df, orders_df, on=["Session ID", "User ID"], how="inner"
)

# Merge the resulting DataFrame with `users_df` on `User ID`
merged_df = pd.merge(merged_df, users_df, on="User ID", how="inner")

# Display the first 5 rows of the merged DataFrame
print("Merged DataFrame Head:")
print(merged_df.head().to_markdown(index=False, numalign="left", stralign="left"))

print("\nMerged DataFrame Info:")
print(merged_df.info())
```

```python
import altair as alt

# Calculate the count of orders for each `Dish Name`
orders = merged_df["Dish Name_x"].value_counts()

# Get the top 5 most ordered dishes
top_dishes = orders.head(5).index.tolist()

# Create a DataFrame for the top dishes
top_dishes_df = merged_df[merged_df["Dish Name_x"].isin(top_dishes)]

# Create a stacked bar chart
chart = (
    alt.Chart(top_dishes_df)
    .mark_bar()
    .encode(
        x=alt.X("Dish Name_x", axis=alt.Axis(title="Dish Name")),
        y=alt.Y("count()", title="No. of Orders"),
        color=alt.Color("Meal Type_x", legend=alt.Legend(title="Meal Type")),
        tooltip=["Dish Name_x", "Meal Type_x", "count()"],
    )
    .properties(title="Top 5 Most Ordered Dishes")
    .interactive()
)

# Display the plot
chart.save("top_5_most_ordered_dishes_by_meal_type.json")
chart.display()
```
